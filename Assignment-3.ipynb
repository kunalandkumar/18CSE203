```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
df=pd.read_csv("https://raw.githubusercontent.com/kunalandkumar/18CSE203/main/the_grammy_awards.csv")
```

```python
plt.figure(figsize=(6,3))
sns.countplot(df.year)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df.year)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='year',y='year',data=df)
plt.show()
```

```python
#correlation
cor=df.corr()
#heatmap for visualising correlation
plt.figure(figsize=(10,8))
sns.heatmap(cor,annot=True,cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(df['year'])
plt.show()
```
